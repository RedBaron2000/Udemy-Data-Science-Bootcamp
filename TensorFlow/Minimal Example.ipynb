```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

### Data generation

```python
observations = 1000

xs = np.random.uniform(low=-10,high=10, size = (observations,1))
zs = np.random.uniform(-10,10,(observations,1))

generate_inputs = np.column_stack((xs,zs))

noise = np.random.uniform(-1,1,(observations,1))

generate_targets = 2*xs - 3*zs + 5 + noise

np.savez('Tf_intro', inputs=generate_inputs, targets=generate_targets)
# file name, label=array targets = arrary
```

### Solving with TensorFlow

```python
training_data = np.load('TF_intro.npz')
```

```python
input_size = 2
output_size = 1

model = tf.keras.Sequential([
                            tf.keras.layers.Dense(output_size)
                            ])

model.compile(optimizer='sgd', loss='mean_squared_error')

model.fit(training_data['inputs'], training_data['targets'], epochs=100, verbose=2)
```

```python
#verbose is whether or not to display output of training 
#value of 2 gives each epoch as 1 line
```

### Extract the weights and bias

### Extract the outputs (make predictions)

```python
model.predict_on_batch(training_data['inputs']).round(1)
```

```python
training_data['targets'].round(1)
```

### Plotting the data

```python
plt.plot(np.squeeze(model.predict_on_batch(training_data['inputs'])), np.squeeze(training_data['targets']))
plt.xlabel('outputs')
plt.ylabel('targers')
plt.show()
```

# customizing TF

```python
input_size = 2
output_size = 1

model = tf.keras.Sequential([
                            tf.keras.layers.Dense(output_size,
                                                 kernel_initializer=tf.random_uniform_initializer(minval=-0.1, maxval=0.1),
                                                 bias_initializer=tf.random_uniform_initializer(minval=-0.1, maxval=0.1)
                                                 )
                            ])

custom_optimizer = tf.keras.optimizers.SGD(learning_rate=0.02)

model.compile(custom_optimizer, loss='mean_squared_error')

model.fit(training_data['inputs'], training_data['targets'], epochs=100, verbose=2)
```
